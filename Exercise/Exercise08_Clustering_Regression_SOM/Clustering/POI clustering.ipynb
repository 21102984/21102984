```python
import numpy as np
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
from matplotlib import pyplot
import pandas as pd
import csv
import os
os.chdir("C:/Users/syq65/Desktop")
```

```python
df = pd.read_csv('POI.csv')
x = df
x_np = np.array(x)  
```

```python
n_clusters = 5                 
estimator = KMeans(n_clusters) 
estimator.fit(x)
```

```python
markers = ['*', 'v', '+', '^', 's','o','x']      # List of markup styles
colors = ['r', 'g', 'm', 'c', 'orange','b','pink']  # List of marker colors
labels = estimator.labels_      # Get cluster labels
centers = estimator.cluster_centers_

pyplot.figure(figsize=(12, 8))
pyplot.title("Clustering of POI in La defense ", fontsize=25)

pyplot.xlabel('Longitude', fontsize=18)
pyplot.ylabel('Latitude', fontsize=18)

for i in range(n_clusters):    # Iterate over all coordinate points and draw scatter plots
    members = labels == i      # Members is a Boolean array  
    pyplot.scatter(
        x_np[members, 1],      # An array of longitude
        x_np[members, 0],      # An array of latitude
        marker = markers[i],   # Mark style
        c = colors[i]          # Mark color
    )   
pyplot.scatter(centers[:,1],centers[:,0],c = 'black',s=150,marker = 'D')
    
pyplot.grid()
pyplot.show()
```
